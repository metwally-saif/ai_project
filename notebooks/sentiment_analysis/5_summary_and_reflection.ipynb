## Performance Overview

### Model 1: Convolutional Neural Network (CNN)
- **Accuracy:** 78.31%
- **Class Performance:**
  - Bearish: Precision 59%, Recall 60%, F1-Score 60%
  - Bullish: Precision 70%, Recall 62%, F1-Score 66%
  - Neutral: Precision 85%, Recall 88%, F1-Score 86%

### Model 2: Bidirectional LSTM (Bi-LSTM)
- **Accuracy:** 76.05%
- **Class Performance:**
  - Bearish: Precision 52%, Recall 58%, F1-Score 54%
  - Bullish: Precision 68%, Recall 54%, F1-Score 60%
  - Neutral: Precision 84%, Recall 87%, F1-Score 85%

### Key Observations:
1. **Overall Accuracy:**
   - Model 1 achieved slightly higher accuracy (78.31%) compared to Model 2 (76.05%).

2. **Class Imbalance:**
   - Both models performed significantly better on the Neutral class, likely due to its larger representation in the dataset (775 samples out of 1194).

3. **Bearish and Bullish Classes:**
   - Both models struggled with the Bearish and Bullish classes, as shown by their lower precision and recall scores.
   - Model 1 had better performance for Bullish sentiment, while Model 2 showed better recall for Bearish sentiment.

---


```python
import matplotlib.pyplot as plt

# Define accuracies
accuracies = [78.31, 76.05]
model_names = ["Model 1 (CNN)", "Model 2 (Bi-LSTM)"]

# Plot accuracies
plt.figure(figsize=(8, 6))
plt.bar(model_names, accuracies, color=["skyblue", "lightgreen"])
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy (%)")
plt.ylim(0, 100)
plt.show()
```

## Observations and Insights

### Model Strengths
1. **Neutral Sentiment:**
   - Both models performed well on the Neutral class, with F1-scores above 85%.
   - The larger representation of Neutral samples likely contributed to this strong performance.

2. **Model 1 (CNN):**
   - Achieved better precision and recall for the Bullish class, making it more suitable for scenarios where Bullish predictions are critical.

3. **Model 2 (Bi-LSTM):**
   - Demonstrated better recall for the Bearish class, which could be beneficial in tasks where identifying bearish trends is crucial.

### Model Weaknesses
1. **Bearish and Bullish Sentiments:**
   - Both models struggled with these classes, likely due to their smaller representation in the dataset (179 and 240 samples, respectively).
   - This imbalance affected the models' ability to generalize.

2. **Overlapping Features:**
   - Sentiment classes may have overlapping features, making it harder for the models to distinguish between Bearish and Bullish sentiments.
